b. Construct the analysis-of-variance table and test for significance of regression.


```python
b0 = -1.8084
b2 = 0.0036
b7 = 0.1940
b8 = -0.0048
```

```python
#Degree of freedom
n = len(df)
df_x1 = 3
df_err = len(df) - 4
df_total = len(df) - 1
print(df_x1, df_err, df_total)
```

```python
#Sum of Squares
SSR = sum(((b0 + b2*df['x2'] + b7*df['x7'] + b8*df['x8']) - np.mean(df['y']))**2)
SSE = sum((df['y'] - (b0 + b2*df['x2'] + b7*df['x7'] + b8*df['x8']))**2)
SSTO = sum((df['y'] - np.mean(df['y']))**2)
print(SSR, SSE, SSTO)
```

```python
#Mean Squares
MSR = SSR / df_x1
MSE = SSE / df_err
print(MSR, MSE)
```

$$H0:\beta_0=\beta_2=\beta_7=\beta_8=0$$
$$Ha: \text{all betas are not 0}$$

```python
#Test for significant of regression
F_star = (SSR/3)/(SSE/(n-4))
print("F* = " + str(F_star))
F = stats.f.ppf(q = 1-0.05, dfn=1, dfd = n-4)
p = 1- stats.f.cdf(x = F_star, dfn=1, dfd = n-4)
print(F,F_star, p)
```

**Conclusion:** F* > F so we conclude Ha, which means that the model is significant

c. Calculate t statistics for testing the hypotheses H0: ùõΩ2 = 0, H0: ùõΩ7 = 0, and H0: ùõΩ8 = 0. What conclusions can you draw about the roles the variables x2, x7, and x8 play in the model?


```python
var_covar_matrix = MSE*(inv_XtX)

s_b2 = np.sqrt(var_covar_matrix[1,1])
s_b7 = np.sqrt(var_covar_matrix[2,2])
s_b8 = np.sqrt(var_covar_matrix[3,3])

t2_star = b2/s_b2
t7_star = b7/s_b7
t8_star = b8/s_b8

print("t2* = " + str(t2_star))
print("t7* = " + str(t7_star))
print("t8* = " + str(t8_star))
```

```python
t_val = stats.t.ppf(q = 1-0.05/2, df=n-4)
print("t = " + str(t_val))
```

**Conclusion:**  
- |t2*| > t so we conclude $$Ha:\beta2\neq0$$
- |t7*| > t so we conclude $$Ha:\beta7\neq0$$
- |t8*| > t so we conclude $$Ha:\beta8\neq0$$  
Therefore, x2, x7 and x8 play a significant role in the model

e. Using the partial F test, determine the contribution of x7 to the model. How is this partial F statistic related to the t test for Œ≤7 calculated in part c above?  
$$H_0: x_7 = 0$$
$$H_a: x_7\neq0$$

```python
x_reduce_raw = df[['x2', 'x8']]
x_reduce = np.hstack((np.ones((n,1)), x_reduce_raw))
XtX_reduce = x_reduce.T @ x_reduce
inv_XtX_reduce = linalg.inv(XtX_reduce)
beta_hat_reduce = inv_XtX_reduce @ x_reduce.T @ Y
y_hat_reduce = x_reduce @ beta_hat_reduce
residuals_reduce = Y - y_hat_reduce
SSE_reduce = sum(residuals_reduce**2)

q = 1
p = X.shape[1]
F_partial = ((SSE_reduce - SSE)/(q))/(SSE/(n-p))
threshold = stats.f.ppf(q = 1-0.05, dfn=q, dfd=n-p)
print(F_partial, threshold)
```

**Conclusion:** F* > F so we conclude Ha, which mean x7 plays a significant role in the model
